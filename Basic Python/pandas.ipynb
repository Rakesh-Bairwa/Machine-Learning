```python
"""METHOD 2 : input some data in nan
1. Put mean
2. Put zero
3. Put most occurening element
4. mean of simliar flower """
df.pl.fillna(df.pl.mean(), inplace = True) # 1, pl = col_name
# creating new data frame
a = df[df.flower == "Setosa"]
df.pw.fillna(a.pw.mean(), inplace = True) # 4, pw = col_name 
df.head()
```

```python
""" working with digits is easiler than, string
so we can change string, ex : male = 0, female = 1"""
#Insert column as set F. to all
df['GENDER'] = "female"
df.head()
```

```python
#Making some GENDER = Male
df.iloc[0:2, 5] = "Male" # iloc[range, col_name]
df.head()
```

```python
def func(s):
    if s == "Male":
        return 0
    else:
        return 1

df['Sex'] = df.GENDER.apply(func) 
"""Take each entry of GENDER col
And apply function : func, and after that store the
Result in another column"""
df.head()
```
