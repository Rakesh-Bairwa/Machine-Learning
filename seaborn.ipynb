```python
# It made on top of matplotlib => seaborn have matplotlib functionality
# It's statistical plotting & vizulization library
# It works pretty good with pandas dataframe
```

```python
import seaborn as sns
import numpy as np
```

```python
# Loading dataset in seaborn
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# BAR GRAPH
sns.barplot(x = 'sex', y = 'total_bill', data = tips) # DEFAULT : estimator is "mean"
```

```python
# Changing estimator form mean to standard devitation in bar graph
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)
```

```python
# Count-plot : count the occurence
# It doesn't require x, y axis just pass the attribute & dataset
sns.countplot(x = 'sex', data = tips) # It doen't have estimator function
```

```python
# BOX-PLOT : distrubution of categorical data
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue ="smoker")
```

```python
# It represent same time like boxplot w/o split it looks simlar to above, in boxplot we don't have split
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue ="smoker", split = True)
```

```python
# Distrubution plot : helps in plotting distrubution of a variable 
sns.distplot(tips["total_bill"]) # It's an univarte distrubution
```

```python
# We can remove the curve using "kde = false", where kde is kernal density estimation
sns.distplot(tips["total_bill"], kde = False, bins = 30) # we can also use bins
```

```python
# We can also only plot kde graph
sns.kdeplot(tips["total_bill"])
```

```python
# JOINT-PLOT : combining two graphs(or variable) together, it's a bi-varate plot
sns.jointplot(x = "total_bill", y = "tip", data = tips)
# histogram above dot graph is made corresponds to x-axis, whereas histogram made vertically corresponds to y-axis
# there is a parameter "kind = scatter", which implies kind of data points inside graph, we've other options 
# like : hex, kde, reg, ...
```

```python
# Unlike joint-plot that plot 2 varibles, we can plot all variables at a time paiwise
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex')
```

```python
tips.corr() # Takes all non categorical data & find corr among them
# In below table we variable on both the axes
```

```python
# HEAT-MAP : only used when we've variables on both the axes of a table
sns.heatmap(tips.corr(), annot = True) # annot will display corr value for that box
#here light color => high corr, whereas dark color => low corr
```

```python
# We can create table like corr using pivot table
data = tips.pivot_table(index = "time", columns = "day", values = "tip")
# it will create a 2d table where for the x = time & y = day. what would be the tip
```

```python
print(data)
```

```python
sns.heatmap(data) # here color =/> corr, but just values from pivot table
```
