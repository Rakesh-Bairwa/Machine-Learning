```python
"""
                        ABSTRACT IDEA
Bar graph : used on categorical data, and how this categories comparing each other
Bar & line graph both used for see trend b/w category's but when few cateorgy it makes more
sense to use Bar, whereas for more category line looks better(ex-stock price) 
Scatter plot is useful when we want to look relationship b/w variables. 
A pie chart useful when we want to something like %age og areas
Bubble plot is used when we've 3-variables, so w/o making graph 3D so, that we can plot 3 varibles on 2D
Histogram is used for frequency (ex-freq distrubution)
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for a 2D graph we need x & y of same size
x = [1, 2, 3]
y = [2, 4, 6]
plt.scatter(x, y)
plt.show()
```

```python
#straight line graph
plt.plot(x , y)
plt.show()
```

```python
#connect points via line
plt.scatter(x, y)
plt.plot(x , y)
plt.show()
```

```python
#ploting 2 line in a graphs
x2 = [3, 4, 5]
y2 = [2, 4, 6]
plt.plot(x2, y2, "r--") # color = red, line --
#single '-' for simple line
# 1st ".." will be "color" & other is graph "marker"
plt.plot(x, y, "go")# color = green, line o o o
plt.show()
```

```python
#plot y = x^3 graph
x = np.arange(0, 5, 0.1) #using small step => smooth graph
y = x**3
plt.plot(x, y)
plt.show()
```

```python
#ploting single array
x = [10, 20, 30, 40]
plt.plot(x, "go", label = "point") # It will take 'x' as 'y-axis' & x-axis
plt.legend()#will be [0,1,2...n] where n = no. of element in x
plt.show()
""" parameter of plot :
linewidht = int, color = "red", marker = "o"
xlabel = "x^3", ylabel, title = "Cubic"
plt.legend() <-- box label at top
"""
```

```python
# we can modify axis lenght : plt.axis(x_start_int, x_end, y_start, y_end)
# adding grid : plt.grid()
# Adding random text on graph : plt.text(x, y, "text_to_show", fontsize,..)
x = [1, 2, 3]
y = [2, 3, 4]
plt.plot(x, y)
plt.text(2, 3, "hello")
plt.grid()
plt.show()
```

```python
# BUBBLE CHART : It's point size can of different size, & size of bubble describe something
# Ex: chart b/w avg salary & year. we can use bubble to denote popultaion of workers
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12,13,14,17,19,20]
population = [100, 120, 180, 250, 300, 370]
#This also works colors = ['r','b','g','y','r','g'];
colors = np.arange(len(year)) # array of 0, 1,..n(total year entries)
plt.scatter(year, salary, s= population, c= colors, alpha = 0.57, edgecolor = 'b')
# s= size, so for each point different size will be taken
# alpha = transparency belong to [0, 1], edgecolor of point
plt.show()
```

```python
#text for each point using loop
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12,13,14,17,19,20]
population = [100, 120, 180, 250, 300, 370]
plt.scatter(year, salary)
for i in [0, 1, 2, 3, 4 ,5]:
    plt.text(year[i] + 0.1, salary[i] + 0.1, population[i])
# This '0.1' help in plotting text little away from point
plt.show()
```

```python
# Pie Chart
sizes = [3,4,6,2]
# If we need to pass label to each section
labels = ["A", "B", "C", "D"]
# If need user defined colors pass colors
colors = ["blue", "purple", "red", "pink"] #diff colors diff areas
# If we want to give our graph a title
plt.title("PIE GRAPH")
plt.pie(sizes, colors = colors, labels = labels)
plt.show()
```

```python
# If we want to enter data in each section => we can make a function or use "autopcf" for %age area
# If we want to label as 26.7% then use autopct = "%.1f%%", here %% is special char, single % after .1f => ERROR, 
plt.pie(sizes, labels = labels, autopct = "%.1f")
plt.show()
plt.pie(sizes, labels = labels, autopct = "%.1f%%")
plt.show()
```

```python
# We can highlight any part of pie with "explode" function : It will detetch that part
sizes = [3,4,6,2]
explode = [0.1, 0.2, 0, 0]
plt.pie(sizes, explode = explode)
plt.show()
```

```python
# Upside down our graph, s.t operation will look clockwise, instead of ant-clockwise as above for explode
plt.pie(sizes, explode = explode, counterclock = False)
plt.show()
```

```python
# unless upside down using counterclock we can rotate our graph acc. to our need
plt.pie(sizes, explode = explode, startangle = 100) # starr from 100 degree
# Before it start from 0 degree --> obserse blue area i.e; 0th index data
plt.show()
```

```python
# HISTOGRAM
a = [2,3,4,1,1,3,2,4,4,4,4,4,4,2]
plt.hist(a)
plt.show()
```

```python
import numpy as np
arr = [0,1,2,2,4,5,5,7,7,8,9,10,10,13,14]
plt.hist(arr, edgecolor = 'black')
plt.show()
print(arr)
```

```python
# xticks : 
axis = np.arange(15)
plt.hist(arr, edgecolor = 'black')
plt.xticks(axis) # divide hist graph into 10(by default) bins(or pieces)
# (max_min - min_data) / total_bins, ex- for bin of widht : 1.7 how many blocks b/w(1-1.7) 
plt.show()
# as can be seen max_data = 14, min_data = 0, by default bin = 10, then each bins width will be
# ((14 - 0) / 10) = 1.4 
```

```python
# passing user defined bins
plt.hist(arr, bins = 3, edgecolor = 'black')
# OR
plt.hist(arr, bins = 14, edgecolor = 'black')

plt.xticks(axis)
plt.show()

# bins data in bars are [a,b) i.e for bin width = 0.9, here x = 1 willn't be included 
# But the last bin is [a,b] => it contain both boundaries
```

```python
# we can provide costum bin 
plt.hist(arr, bins = [1,3,6,11,14], edgecolor = 'black')
# It will make 4 bins as [1,3),[3,6),[6,11),[11,14]
plt.xticks(axis)
plt.show()
```

```python
# Providing X & y axis range
arr = [0,1,2,2,4,5,5,7,7,8,9,10,10,13,14]
plt.hist(arr, edgecolor = 'black')
plt.axis([0,20, 0, 6]) # [x_min, x_max, y_min, y_max]
plt.show()
```

```python
# BAR GRAPHS
year = [2015, 2016, 2017, 2018, 2019, 2020]
population = [100, 200, 300, 400, 500, 600]
plt.bar(year, population)
plt.show()
```

```python
# we can change it's width of bars, we can rotate text below bars using "xticks"
plt.bar(year, population, width = 0.4)
plt.xticks(rotation = 45)
plt.show()
# Simlarily we can add more attributes to our bar graph
```
